Identified Data Inconsistencies & Fixes
1. Yam Tuber (yam_tuber)

Observation: Prices drop from ~â‚¦48,000 (2012â€“2015) to ~â‚¦500 (2016â€“Present).

Diagnosis: Unit error. Early data tracks 100kg Bags; later data tracks Single Tubers.

Fix: Divide pre-2016 prices by 100.

2. Garri White (garri_white)

Observation: Massive price spike to ~â‚¦7,000 in Jan 2020 (vs. typical â‚¦300).

Diagnosis: Unit mixing. The Jan 2020 entries record 50kg Bag prices in the "Paint Bucket" column.

Fix: Divide any price > â‚¦2,000 by 25 to normalize to bucket size.

3. Rice Local (rice_local)

Observation: Deflationary trend where 2015 prices (~â‚¦17,500) are higher than 2018 prices (~â‚¦7,000).

Diagnosis: Unit error. 2015 prices likely reflect 100kg/Double Bags, while 2018 reflects standard 50kg Bags.

Fix: Divide pre-2017 prices > â‚¦13,000 by 2.

4. Palm Oil (palm_oil)

Observation: Prices drop from ~â‚¦31,000 (2014â€“2015) to ~â‚¦2,600 (2018).

Diagnosis: Unit error. Early data tracks 25L/50L Kegs, later data tracks 5L Kegs.

Fix: Divide pre-2017 prices by 10 (or 12) to align with the 5L unit.

5. Bread Loaf (bread_loaf)

Observation: Illogical price crash from ~â‚¦600 (2019) to ~â‚¦300 (2021) despite high inflation.

Diagnosis: Item specification change. 2018â€“2020 tracks Jumbo Loaves, while 2021 switches to Small Loaves.

Fix: Multiply 2021â€“2022 prices by 2 to standardize to "Jumbo Loaf" equivalent (or scale down the earlier years). Recommendation: Scale up 2021-2022 to match the inflation trend.

# CLEANING STEP

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Load Data
try:
    # Starting from the raw file to ensure all steps are applied cleanly
    df = pd.read_csv('final_dataset (2).csv')
    df['date'] = pd.to_datetime(df['date'])
    df_clean = df.copy()

    print("STARTING DATA TRANSFORMATION...")

    # ==========================================
    # CLEANING LOGIC (Consolidated)
    # ==========================================

    # 1. DIESEL: Fix Zero/Flatline Errors
    # Replace anything < 50 with NaN (to be filled later)
    mask_diesel_zero = (df_clean['item_id'] == 'diesel') & (df_clean['price_ngn'] < 50)
    df_clean.loc[mask_diesel_zero, 'price_ngn'] = np.nan

    # 2. YAM TUBER: Fix Bag Prices
    # Any Yam > 2500 before 2021 is a Bag error. Divide by 100.
    mask_yam_bag = (df_clean['item_id'] == 'yam_tuber') & (df_clean['price_ngn'] > 2500) & (df_clean['date'].dt.year < 2021)
    df_clean.loc[mask_yam_bag, 'price_ngn'] /= 100

    # 3. PALM OIL: Fix Drum/Jerrycan Prices
    # Any Oil > 8000 is a Drum error. Divide by 15.
    mask_oil = (df_clean['item_id'] == 'palm_oil') & (df_clean['price_ngn'] > 8000)
    df_clean.loc[mask_oil, 'price_ngn'] /= 15

    # 4. GARRI: Fix Bag Spikes
    # Any Garri > 2000 is a Bag error. Divide by 25.
    mask_garri = (df_clean['item_id'] == 'garri_white') & (df_clean['price_ngn'] > 2000)
    df_clean.loc[mask_garri, 'price_ngn'] /= 25

    # 5. RICE LOCAL: Fix Double Bags & Mudus
    mask_rice_high = (df_clean['item_id'] == 'rice_local') & (df_clean['price_ngn'] > 13000) & (df_clean['date'].dt.year <= 2016)
    df_clean.loc[mask_rice_high, 'price_ngn'] /= 2
    mask_rice_low = (df_clean['item_id'] == 'rice_local') & (df_clean['price_ngn'] < 3000)
    df_clean.loc[mask_rice_low, 'price_ngn'] *= 20

    # 6. BREAD: Fix Small Loaf Switch
    mask_bread = (df_clean['item_id'] == 'bread_loaf') & (df_clean['price_ngn'] < 400) & (df_clean['date'].dt.year >= 2021)
    df_clean.loc[mask_bread, 'price_ngn'] *= 2

    # INTERPOLATION WITH INDEX FIX
    df_clean = df_clean.set_index('date')
    df_clean['price_ngn'] = df_clean.groupby('item_id')['price_ngn'].transform(
        lambda x: x.interpolate(method='time').ffill().bfill()
    )
    df_clean = df_clean.reset_index()


    # FINAL MICRO-POLISH

    # 1. Force Diesel Reality: Minimum price is 150
    mask_diesel = df_clean['item_id'] == 'diesel'
    df_clean.loc[mask_diesel, 'price_ngn'] = df_clean.loc[mask_diesel, 'price_ngn'].clip(lower=150)

    # 2. Clip lingering Yam Spike (Lowering threshold to 1500 for early years)
    mask_yam_outlier = (df_clean['item_id'] == 'yam_tuber') & (df_clean['price_ngn'] > 1500) & (df_clean['date'].dt.year < 2017)
    df_clean.loc[mask_yam_outlier, 'price_ngn'] /= 5

    # Save cleaned prices (for visual checking if needed)
    df_clean.to_csv('final_dataset_cleaned_v3.csv', index=False)
    print("Cleaned dataset saved as final_dataset_cleaned_v3.csv")


    # PLOTTING GRAPHS (ALL ITEMS)

    # Get list of unique items
    items = df_clean['item_id'].unique()

    # Determine grid size
    num_items = len(items)
    cols = 2
    rows = (num_items + 1) // cols

    fig, axes = plt.subplots(rows, cols, figsize=(15, 5 * rows))
    axes = axes.flatten()

    for i, item in enumerate(items):
        subset = df_clean[df_clean['item_id'] == item]
        axes[i].plot(subset['date'], subset['price_ngn'], label=item, linewidth=1.5)
        axes[i].set_title(f"Price Trend: {item}")
        axes[i].set_ylabel("Price (NGN)")
        axes[i].legend()
        axes[i].grid(True, alpha=0.3)

    # Hide empty subplots if any
    for j in range(i + 1, len(axes)):
        axes[j].axis('off')

    plt.tight_layout()
    plt.show()


    # PHASE 3: TRANSFORMATION & FEATURE ENGINEERING

    print("Pivoting data to Wide Format...")
    df_pivot = df_clean.pivot_table(index='date', columns='item_id', values='price_ngn')

    # Resample to Daily
    df_pivot = df_pivot.resample('D').ffill()

    # Calculate Log Returns
    print("Calculating Daily Returns (Momentum)...")
    df_returns = np.log(df_pivot / df_pivot.shift(1))

    # Replace infinite values and NaNs
    df_returns.replace([np.inf, -np.inf], 0, inplace=True)
    df_returns.fillna(0, inplace=True)

    # Save final training data
    df_returns.to_csv('final_training_data.csv')
    print("SUCCESS: Created final_training_data.csv")
    print(f"Dimensions: {df_returns.shape}")

except Exception as e:
    print(f"Error: {e}")
```

```python
def create_lagged_features(data, target_cols, driver_cols, lags):
    """
    Creates lagged columns for specified drivers and targets.

    Args:
        data: DataFrame of daily returns
        target_cols: List of items we want to predict (e.g., Rice, Garri)
        driver_cols: List of items that drive inflation (e.g., Fuel, Diesel)
        lags: List of days to look back (e.g., [1, 7, 14])

    Returns:
        DataFrame with new lagged columns, NaN rows dropped.
    """
    df_lags = data.copy()

    # 1. Generate Lags for Drivers (External Factors)
    # Example: Fuel_Lag_7 (Fuel price 7 days ago)
    for col in driver_cols:
        if col in df_lags.columns:
            for lag in lags:
                new_col_name = f"{col}_lag_{lag}"
                df_lags[new_col_name] = df_lags[col].shift(lag)

    # 2. Generate Lags for Targets (Autoregression)
    # Example: Rice_Lag_1 (Rice price yesterday predicting Rice price today)
    for col in target_cols:
        if col in df_lags.columns:
            for lag in lags:
                new_col_name = f"{col}_lag_{lag}"
                df_lags[new_col_name] = df_lags[col].shift(lag)

    # 3. Drop rows with NaNs created by shifting
    # (The first 30 days will have missing data because they have no "past")
    df_lags.dropna(inplace=True)

    return df_lags
```

```python
# Define Drivers and Targets
# Drivers: External costs that push prices up
drivers = ['fuel', 'diesel', 'palm_oil']

# Targets: The food items we want to predict
targets = ['rice_local', 'rice_foreign', 'garri_white', 'yam_tuber', 'bread_loaf', 'beans_oloyin']

# Define the Time Windows (Lags)
# 1 Day = Immediate Momentum
# 7 Days = Weekly Cycle
# 14 Days = Bi-weekly Cycle
# 30 Days = Monthly Cycle
lag_periods = [1, 7, 14, 30]

# Execute Feature Engineering
df_features = create_lagged_features(df, targets, drivers, lag_periods)

# Save the final dataset for MLOps/Modeling
df_features.to_csv('final_model_ready_data.csv')

print("Feature Engineering Complete.")
print(f"Original Dimensions: {df.shape}")
print(f"New Dimensions: {df_features.shape}")
print(f"Total Features Created: {df_features.shape[1]}")
print("File saved: 'final_model_ready_data.csv'")
```

```python
from datetime import datetime

class ExperimentTracker:
    def __init__(self, log_file='model_leaderboard.csv'):
        self.log_file = log_file

        # Initialize the log file if it doesn't exist
        if not os.path.exists(self.log_file):
            df = pd.DataFrame(
                columns=[
                    'timestamp',
                    'model_name',
                    'target_item',
                    'features_used',
                    'rmse_score',
                    'mae_score',
                    'r2_score',
                ]
            )
            df.to_csv(self.log_file, index=False)
            print(f"Initialized new Experiment Log: {self.log_file}")
        else:
            print(f"Found existing Experiment Log: {self.log_file}")

    def log_experiment(self, model_name, target, features, y_true, y_pred):
        """
        Logs the results of a single training run.
        """
        # Calculate Metrics
        rmse = np.sqrt(mean_squared_error(y_true, y_pred))
        mae = mean_absolute_error(y_true, y_pred)
        r2 = r2_score(y_true, y_pred)

        # Create Log Entry
        entry = {
            'timestamp': datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            'model_name': model_name,
            'target_item': target,
            'features_used': len(features),
            'rmse_score': round(rmse, 5),  # Lower is better
            'mae_score': round(mae, 5),  # Lower is better
            'r2_score': round(r2, 5),  # Higher is better (1.0 is perfect)
        }

        # Append to CSV
        df_entry = pd.DataFrame([entry])
        df_entry.to_csv(self.log_file, mode='a', header=False, index=False)

        print(f"Logged: {model_name} | RMSE: {rmse:.4f}")
        return rmse

    def print_leaderboard(self):
        """
        Displays the current rankings of all models.
        """
        if os.path.exists(self.log_file):
            df = pd.read_csv(self.log_file)
            if df.empty:
                print("Leaderboard is empty.")
                return

            # Sort by RMSE (Ascending - Lower error is better)
            leaderboard = df.sort_values(by='rmse_score', ascending=True)
            print("\n=== MODEL LEADERBOARD ===")
            print(leaderboard[['model_name', 'target_item', 'rmse_score', 'r2_score']].to_string(index=False))
        else:
            print("No log file found.")

# Initialize the Tracker
tracker = ExperimentTracker()
```

```python
from sklearn.linear_model import ElasticNetCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
targets = ['rice_local', 'garri_white', 'yam_tuber', 'bread_loaf', 'beans_oloyin']
features = [col for col in df.columns if col not in targets]

print(f"Data Loaded. Training Start Date: {df.index.min()}")
print(f"Testing Start Date (approx): {df.index[-int(len(df)*0.2)]}")
```

```python
# Re-initialize Tracker (in case cell was not run)
if 'tracker' not in locals():
    # Simple placeholder if class is missing from memory
    print("Tracker not found in memory. Printing results to screen instead.")
    class SimpleTracker:
        def log_experiment(self, name, target, feats, y_true, y_pred):
            rmse = np.sqrt(mean_squared_error(y_true, y_pred))
            print(f"Logged: {name} ({target}) | RMSE: {rmse:.4f}")
    tracker = SimpleTracker()
```

```python
# 4. Training Loop
results = {} # Initialize results dictionary outside the loop

for target in targets:
    print(f"\n--- Training Elastic Net for {target} ---")

    # Select features: All numeric columns except the target itself (and its future values)
    # We strictly filter for numeric types to avoid the "String to Float" error
    numeric_df = df.select_dtypes(include=[np.number])

    # Drop the target column from X (we want to predict it, not use it)
    # Exclude 'Unnamed: 0' which is often an artifact of saving/loading CSVs
    feature_cols = [c for c in numeric_df.columns if c != target and c != 'Unnamed: 0']

    X = numeric_df[feature_cols]
    y = numeric_df[target]

    # 2. Time-Series Split (80% Train, 20% Test)
    # CRITICAL: We do not shuffle. We slice by time.
    split_point = int(len(df) * 0.8)
    X_train, X_test = X.iloc[:split_point], X.iloc[split_point:]
    y_train, y_test = y.iloc[:split_point], y.iloc[split_point:]

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Train the model
    model = ElasticNetCV(l1_ratio=[.1, .5, .9, .99], cv=5, random_state=42)
    model.fit(X_train_scaled, y_train)

    # Predict
    y_pred = model.predict(X_test_scaled)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mae = mean_absolute_error(y_test, y_pred)

    print(f"Model Trained. RMSE: {rmse:.5f}")
    print(f"Model Trained. MAE: {mae:.5f}")

    # Store results
    results[target] = {
        'model': model,
        'rmse': rmse,
        'coefs': model.coef_,
        'features': feature_cols,
        'y_test': y_test,
        'y_pred': y_pred
    }

    # Log experiment with the tracker
    tracker.log_experiment(
        model_name='ElasticNetCV',
        target=target,
        features=feature_cols,
        y_true=y_test,
        y_pred=y_pred
    )
```

# DEEPLEARNING REQUIRES A 3D DATA SHAPE (Samples, Time Steps, Features)

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# 1. Load the Returns Data (Momentum)
# We go back to the Returns data because it is stationary (trainable).
df = pd.read_csv('final_training_data.csv', index_col=0, parse_dates=True)
targets = ['rice_local', 'garri_white', 'yam_tuber', 'bread_loaf', 'beans_oloyin']

print(f"Data Loaded. Shape: {df.shape}")

def get_lstm_data(df, target_col, look_back=30):
    """
    Creates 3D sequences for LSTM: (Samples, 30 Days, Features)
    """
    data = df.values

    # We want to predict the Target Column
    target_idx = df.columns.get_loc(target_col)

    # Generator produces batches of (Past 30 Days -> Next Day Target)
    gen = TimeseriesGenerator(
        data,
        data[:, target_idx],
        length=look_back,
        batch_size=32
    )
    return gen

print("LSTM Setup Complete. Ready to build the brain.")
```

```python
# Configuration
LOOK_BACK = 30  # Look at the past 30 days
EPOCHS = 20     # How many times to practice
results_lstm = {}

print("=== ROUND 4: LSTM DEEP LEARNING ===")

for target in targets:
    print(f"\n Training LSTM for {target}")

    # 1. Split Data (Keep Order!)
    train_size = int(len(df) * 0.8)
    train_df = df.iloc[:train_size]
    test_df = df.iloc[train_size:]

    # 2. Scale Data (Crucial for Neural Networks)
    # LSTM works best when numbers are between -1 and 1
    scaler = MinMaxScaler(feature_range=(-1, 1))
    train_scaled = scaler.fit_transform(train_df)
    test_scaled = scaler.transform(test_df) # Use same scaler

    # 3. Create Sequences
    # We create a 3D structure: [Samples, 30_Days, 55_Features]
    train_gen = TimeseriesGenerator(
        train_scaled,
        train_scaled[:, df.columns.get_loc(target)],
        length=LOOK_BACK,
        batch_size=16
    )
    test_gen = TimeseriesGenerator(
        test_scaled,
        test_scaled[:, df.columns.get_loc(target)],
        length=LOOK_BACK,
        batch_size=1
    )

    # 4. Build Model
    model = Sequential()
    model.add(LSTM(50, activation='tanh', input_shape=(LOOK_BACK, df.shape[1])))
    model.add(Dropout(0.2)) # Randomly forget 20% to prevent memorization
    model.add(Dense(1))     # Output: Predicted Return

    model.compile(optimizer='adam', loss='mse')

    # 5. Train
    # verbose=0 keeps it clean, change to 1 to see progress bars
    history = model.fit(train_gen, epochs=EPOCHS, verbose=0)

    # 6. Evaluate
    # We predict on the test set
    predictions = model.predict(test_gen, verbose=0)

    # Align Truth (y_true) with Predictions
    # The generator eats the first 30 days, so y_true starts at index 30
    y_true = test_scaled[LOOK_BACK:, df.columns.get_loc(target)]

    # Calculate Scores
    rmse = np.sqrt(mean_squared_error(y_true, predictions))
    r2 = r2_score(y_true, predictions)

    print(f" RMSE: {rmse:.5f} | R2: {r2:.5f}")

    # 7. Log to Tracker
    if 'tracker' in locals():
        tracker.log_experiment(
            model_name="LSTM_DeepNet",
            target=target,
            features=["All_55_Items_Lag_30"],
            y_true=y_true,
            y_pred=predictions.flatten()
        )

    # Store for Plotting
    results_lstm[target] = {
        'y_true': y_true,
        'y_pred': predictions.flatten()
    }

# Final Leaderboard
if 'tracker' in locals():
    tracker.print_leaderboard()
```

```python
import logging
from sklearn.preprocessing import MinMaxScaler

# Suppress Prophet logs
logging.getLogger('cmdstanpy').setLevel(logging.WARNING)

# 1. DATA LOADING & CLEANING
df = pd.read_csv('final_dataset (2).csv')
df['date'] = pd.to_datetime(df['date'])
df_clean = df.copy()

# Apply Surgical Fixes
mask_diesel_zero = (df_clean['item_id'] == 'diesel') & (df_clean['price_ngn'] < 50)
df_clean.loc[mask_diesel_zero, 'price_ngn'] = np.nan
mask_yam_bag = (df_clean['item_id'] == 'yam_tuber') & (df_clean['price_ngn'] > 2500) & (df_clean['date'].dt.year < 2021)
df_clean.loc[mask_yam_bag, 'price_ngn'] /= 100
mask_oil = (df_clean['item_id'] == 'palm_oil') & (df_clean['price_ngn'] > 8000)
df_clean.loc[mask_oil, 'price_ngn'] /= 15
mask_garri = (df_clean['item_id'] == 'garri_white') & (df_clean['price_ngn'] > 2000)
df_clean.loc[mask_garri, 'price_ngn'] /= 25
mask_rice_high = (df_clean['item_id'] == 'rice_local') & (df_clean['price_ngn'] > 13000) & (df_clean['date'].dt.year <= 2016)
df_clean.loc[mask_rice_high, 'price_ngn'] /= 2
mask_rice_low = (df_clean['item_id'] == 'rice_local') & (df_clean['price_ngn'] < 3000)
df_clean.loc[mask_rice_low, 'price_ngn'] *= 20
mask_bread = (df_clean['item_id'] == 'bread_loaf') & (df_clean['price_ngn'] < 400) & (df_clean['date'].dt.year >= 2021)
df_clean.loc[mask_bread, 'price_ngn'] *= 2

# Interpolation
df_clean = df_clean.set_index('date')
df_clean['price_ngn'] = df_clean.groupby('item_id')['price_ngn'].transform(lambda x: x.interpolate(method='time').ffill().bfill())
df_clean = df_clean.reset_index()

# Pivot to Wide Format (Prices)
df_prices = df_clean.pivot_table(index='date', columns='item_id', values='price_ngn').resample('D').ffill()

# Calculate Returns (Momentum)
df_returns = np.log(df_prices / df_prices.shift(1)).fillna(0)

# Define Drivers and Targets
drivers = ['fuel', 'diesel', 'palm_oil']
targets = ['rice_local', 'beans_oloyin', 'yam_tuber', 'garri_white', 'bread_loaf']

# 2. FEATURE ENGINEERING (LAGS) for ElasticNet
def create_lags(data, target_cols, driver_cols, lags=[1, 7, 14, 30]):
    df_lags = data.copy()
    for col in driver_cols + target_cols:
        for lag in lags:
            df_lags[f"{col}_lag_{lag}"] = df_lags[col].shift(lag)
    df_lags.dropna(inplace=True)
    return df_lags

df_features = create_lags(df_returns, targets, drivers)

# 3. MODEL TRAINING & COMPARISON LOOP
results_viz = {}

print("Starting Multi-Model Comparison (ElasticNet vs Prophet vs LSTM)...")

for target in targets:
    print(f"Processing {target}...")

    # --- COMMON DATA SPLIT ---
    # We use the last 20% for testing across all models
    test_size = int(len(df_features) * 0.2)
    train_data = df_features.iloc[:-test_size]
    test_data = df_features.iloc[-test_size:]

    # Store Actuals for Plotting
    y_test_actual = test_data[target]
    dates_test = test_data.index

    # --- MODEL 1: ELASTIC NET ---
    # Prepare X and y
    feature_cols = [c for c in df_features.columns if c != target] # Use lags + drivers
    X_train_en = train_data[feature_cols]
    y_train_en = train_data[target]
    X_test_en = test_data[feature_cols]

    # Scale
    scaler_en = StandardScaler()
    X_train_scaled = scaler_en.fit_transform(X_train_en)
    X_test_scaled = scaler_en.transform(X_test_en)

    # Fit
    en = ElasticNetCV(cv=3, random_state=42)
    en.fit(X_train_scaled, y_train_en)
    pred_en = en.predict(X_test_scaled)

    # --- MODEL 2: PROPHET (On Returns) ---
    # Prophet requires 'ds' and 'y' columns
    # We use the returns dataframe directly but align with the feature engineered dates
    prophet_df = df_returns[[target] + drivers].loc[df_features.index].reset_index()
    prophet_df.rename(columns={'date': 'ds', target: 'y'}, inplace=True)

    prophet_train = prophet_df.iloc[:-test_size]
    prophet_test = prophet_df.iloc[-test_size:]

    model_prophet = Prophet(daily_seasonality=False, yearly_seasonality=True)
    for driver in drivers:
        model_prophet.add_regressor(driver)

    model_prophet.fit(prophet_train)

    # Predict (Prophet needs future regressor values, which we have in test set)
    forecast = model_prophet.predict(prophet_test.drop(columns=['y']))
    pred_prophet = forecast['yhat'].values

    # --- MODEL 3: LSTM ---
    # LSTM needs sequences. We train on the Returns Data (scaled)
    # Scale Target for LSTM (-1, 1 range)
    scaler_lstm = MinMaxScaler(feature_range=(-1, 1))
    # We scale the whole column to ensure consistency
    target_data = df_returns[target].values.reshape(-1, 1)
    scaled_data = scaler_lstm.fit_transform(target_data)

    # Align with the index used by other models (df_features drops initial NaN rows)
    # We need to map the indices correctly.
    # Simplification: We just use the tail of the full dataset matching the test size.

    look_back = 30

    # Generator for Test Data
    # We need the last 'look_back' points from Train to start predicting the first point of Test
    # Total needed = Test Size + Look Back
    test_input = scaled_data[-(test_size + look_back):]

    test_gen = TimeseriesGenerator(test_input, test_input, length=look_back, batch_size=1)

    # Train Data Generator
    train_input = scaled_data[:-(test_size)]
    train_gen = TimeseriesGenerator(train_input, train_input, length=look_back, batch_size=16)

    # Build LSTM
    model_lstm = Sequential()
    model_lstm.add(LSTM(50, activation='tanh', input_shape=(look_back, 1)))
    model_lstm.add(Dense(1))
    model_lstm.compile(optimizer='adam', loss='mse')

    # Fit (Quick training for demo)
    model_lstm.fit(train_gen, epochs=5, verbose=0)

    # Predict
    pred_lstm_scaled = model_lstm.predict(test_gen, verbose=0)
    pred_lstm = scaler_lstm.inverse_transform(pred_lstm_scaled).flatten()

    # --- STORE RESULTS ---
    # Ensure all lengths match (Prophet and EN match exactly. LSTM might be off by 1-2 due to generator)
    # We trim to the minimum common length for plotting
    min_len = min(len(pred_en), len(pred_prophet), len(pred_lstm))

    results_viz[target] = {
        'dates': dates_test[-min_len:],
        'actual': y_test_actual.values[-min_len:],
        'elastic_net': pred_en[-min_len:],
        'prophet': pred_prophet[-min_len:],
        'lstm': pred_lstm[-min_len:]
    }

# 4. PLOTTING
fig, axes = plt.subplots(len(targets), 1, figsize=(12, 4 * len(targets)))
if len(targets) == 1: axes = [axes] # Handle single target case

plt.subplots_adjust(hspace=0.4)

for i, target in enumerate(targets):
    res = results_viz[target]
    ax = axes[i]

    # Plot only last 100 days for clarity
    plot_len = 100
    dates = res['dates'][-plot_len:]

    ax.plot(dates, res['actual'][-plot_len:], label='Actual Returns', color='black', alpha=0.3, linewidth=2)
    ax.plot(dates, res['elastic_net'][-plot_len:], label='ElasticNet', color='green', linestyle='--', linewidth=1.5)
    ax.plot(dates, res['prophet'][-plot_len:], label='Prophet', color='blue', linestyle='-.', linewidth=1.5)
    ax.plot(dates, res['lstm'][-plot_len:], label='LSTM', color='red', linestyle=':', linewidth=1.5)

    # Calc Metrics for Title
    r2_en = r2_score(res['actual'], res['elastic_net'])
    r2_pr = r2_score(res['actual'], res['prophet'])
    r2_ls = r2_score(res['actual'], res['lstm'])

    ax.set_title(f"{target.upper()} Comparison | R2: EN={r2_en:.2f}, Prophet={r2_pr:.2f}, LSTM={r2_ls:.2f}")
    ax.set_ylabel("Daily Momentum (Log Return)")
    ax.legend(loc='upper right')
    ax.grid(True, alpha=0.3)

plt.show()
```
